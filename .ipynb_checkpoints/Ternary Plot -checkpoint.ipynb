# Ternary Plot 

Ternary plot is an extremely useful visualization in material science to discover potential composition that creates material with enhanced properties. 

For this project, we are interested to find out new composition of glass that leads to enhanced mechanical properties. Glasses are made up of CaO (Calcium oxide), Al2O3 (Aluminium oxide) and SiO2 (Silicon dioxide). Some of our interested mechanical properties include density, Young's modulus, Shear modulus, Bulk modulus, Poisson ratio and fracture toughness, which are good indicator of strenght of the glass.

Now we can proceed to define a function to draw our ternary plot

```python
def drawTernaryPlot(df, col):

    class ColumnExtractor(TransformerMixin, BaseEstimator):
        def __init__(self,cols):
            self.cols = cols
        
        def transform(self,X, **transform_params):
            return X[self.cols].dropna()
    
        def fit(self,X,y=None, **fit_params):
            return self

    colext = ColumnExtractor(['CaO', 'Al2O3', col])
    df_model = colext.fit_transform(df)

    def generate_data(df_model):
        d = dict()
        for z in range(len(df_model)):
            d[(df_model.iloc[z,0]),(df_model.iloc[z,1])] = df_model.iloc[z,2]
        return d

    scale = 100
    d = generate_data(df_model)
    fig, ax = plt.subplots()
    ax.axis("off")
    figure, tax = ternary.figure(ax=ax, scale=scale)
    tax.heatmap(d, scale = 100,style = 'h')

    # Draw Boundary and Gridlines
    tax.gridlines(color="black", multiple=10)

    # Set Axis labels and Title
    fontsize = 20
    tax.set_title("Ternary Plot for " + col, fontsize=fontsize)
    tax.bottom_axis_label("CaO", fontsize=fontsize)
    tax.left_axis_label("Al2O3", fontsize=fontsize)
    tax.right_axis_label("SiO2", fontsize=fontsize)

    # Set ticks
    tax.ticks(multiple =10, axis='lbr', linewidth=1)
```

With this function, we can create a ternary plot for each mechanical properties.

```python
columns = ['molar_volume', 'density', 'youngs_modulus',
       'shear_modulus', 'bulk_modulus', 'poisson_ratio']
```

```python
for col in columns:
    drawTernaryPlot(cas_clean, col)
```
